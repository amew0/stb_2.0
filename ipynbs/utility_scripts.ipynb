# **Remove Corrupt Images**

```python
#remove corrupt images    
dir_path = "dataset/Plastic"
# files = dir_path.glob("*")
count = 0

for folder in os.listdir(dir_path):
  folder_path = os.path.join(dir_path, folder)
  for filename_ in os.listdir(folder_path):
      file = os.path.join(dir_path, folder, filename_)
      try:
        img = Image.open(file)
        img.verify()
      except (IOError,SyntaxError) as e:
        print(str(file))
        os.remove(file)
        count = count + 1
print(count)

```

# **Move Files from folder to another**

```python
def move_files(src,dest):
  old_path = Path(src)
  files = old_path.glob('*/*')
  count = 0
  n = 5
  for index,file in enumerate(files):
   try:
      if(index%n==0):
        shutil.move(str(file),str(dest))
        count = count + 1
      if(count == 78):
        break
   except:
     pass
    
       
     
   
```

# **Extracting images from video input**

```python
# Importing all necessary libraries
import cv2
import os
from datetime import datetime
# Read the video from specified path
cam = cv2.VideoCapture("/content/drive/My Drive/dataset-videos/plastic/28.mp4")

try:
	
	# creating a folder named data
	if not os.path.exists('/content/drive/My Drive/dataset/plastic/28'):
		os.makedirs('/content/drive/My Drive/dataset/plastic/28')

# if not created then raise error
except OSError:
	print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):
	
	# reading from frame
	ret,frame = cam.read()

	if ret:
		# if video is still left continue creating images
		now = datetime.now()
		current_time = now.strftime("%H:%M:%S")
		name = '/content/drive/My Drive/dataset/plastic/28/plastic_' + current_time+str(currentframe) + '.PNG'
		print ('Creating...' + name)

		# writing the extracted images
		cv2.imwrite(name, frame)

		# increasing counter so that it will
		# show how many frames are created
		currentframe += 1
	else:
		break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os
from google.colab.patches import cv2_imshow
folder = pathlib.Path("/content/drive/My Drive/data-camera-rpi/Plastic")
files = folder.glob("*")
count = 0
import numpy as np
for image in files:
  img_bgr = cv2.imread(str(image))
  cv2_imshow(img_bgr)
  #os.remove(image)
  # Convert the image from BGR to RGB format
  img_rgb_ = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) # COLOR_BGR2RGB
  cv2_imshow(img_rgb_)
  count = count + 1

  if np.array_equal(img_bgr, img_rgb_):
    print(True)
  else:
    print(False)
  if count == 10:
    break
```
