## Load The Pre-trained Model - **RESNET50**

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
# Create a function to convert labels to one-hot encoding
num_classes = 3
def to_one_hot(x, y):
    num_classes = 3
    y_one_hot = tf.one_hot(y, num_classes)
    return x, y_one_hot
```

```python
train_ds = train_ds.map(to_one_hot)
test_ds = test_ds.map(to_one_hot)
```

```python
base_model = InceptionV3(input_shape=img_size + (3,),
                         weights='imagenet',
                         include_top=False)
```

```python
# Freeze all layers in the base model
for layer in base_model.layers:
    layer.trainable = False
    
    # print(layer.layers.get_config()['name'])
```

```python
base_model.layers[20].layers.get_config()
```

```python
# block1 = model.layers[1:8]       # Stem
# block2 = model.layers[8:30]      # Inception module 1
# block3 = model.layers[30:55]     # Inception module 2
# block4 = model.layers[55:87]     # Inception module 3
# block5 = model.layers[87:]       # Fully connected layers
for layer in base_model.layers[55:87]:
    layer.trainable=True
```

```python
for layer in base_model.layers:
    if(layer.trainable):
        print(layer.get_config()['name'])
```

```python
# Add new layers on top of the base model
x = tf.keras.layers.Flatten()(base_model.output)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
output = tf.keras.layers.Dense(len(class_names), activation='softmax')(x)
```

```python
# Define the model and compile it
model = tf.keras.Model(inputs=base_model.input, outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
log_dir = "logs/InceptionV3-STBv1.7"

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

chpt = "Checkpoint/InceptionV3-STBv1.7"

# Create the ModelCheckpoint callback
checkpoint_callback = ModelCheckpoint(
    filepath = chpt+'/{epoch:02d}.h5',
    save_weights_only=False,
    save_best_only=False,
    save_freq='epoch',
    verbose=1)
```
